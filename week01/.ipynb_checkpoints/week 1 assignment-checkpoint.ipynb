Because my data does not have latitude or longitude listed, I'm going to try and extract that information from the geometry column. Wish me luck!

```python
buyout_geo_recent['LONG'] = buyout_geo_recent['geometry'].x
buyout_geo_recent['LAT'] = buyout_geo_recent['geometry'].y
```

```python
buyout_geo_recent['LONG']
```

Success! Going to continue on trying to map my data.

```python
latitude = buyout_geo_recent.LAT.mean()
latitude
```

```python
longitude = buyout_geo_recent.LONG.mean()
longitude
```

```python
m = folium.Map(location=[latitude,longitude])
m
```

```python
for index, row in buyout_geo_recent.iterrows():
    print(row.analysis_n, row.LAT, row.LONG)
```

```python
for index, row in buyout_geo_recent.iterrows():
    # add folium marker code
    folium.Marker([row.LAT, row.LONG], popup=row.analysis_n, tooltip=row.analysis_n).add_to(m)
m
```
